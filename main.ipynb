```python
text = """
Python is a programming language that lets you work quickly 
and integrate systems more effectively. Python is popular and easy to learn.
"""
print(text)

```

```python
words = text.split()
words
```

```python
p_words = [w for w in words if w.lower().startswith("p")]
p_words
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

wc = WordCloud(width=800, height=400, background_color="white").generate(text)

plt.figure(figsize=(10,5))
plt.imshow(wc)
plt.axis("off")
```

### Explanation

The word cloud shows which words occur most frequently in the text.
The largest word is "Python" because it occurs several times in the text.
Smaller words occur less frequently.
